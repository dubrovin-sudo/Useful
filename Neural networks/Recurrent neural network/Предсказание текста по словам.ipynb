# Загрузка библиотек

```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '2'

import numpy as np

from tensorflow.keras.layers import Dense, SimpleRNN, Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import Tokenizer, text_to_word_sequence
from tensorflow.keras.utils import to_categorical
```

# Загрузка файла для обучения

```python
with open('chemodan.txt', 'r', encoding='utf-8') as f:
    texts = f.read()
    texts = texts.replace('\ufeff', '')  # убираем первый невидимый символ
```

# Парсинг текста

```python
# парсим текст, как последовательность слов
maxWordsCount = 10000
tokenizer = Tokenizer(num_words=maxWordsCount, filters='!–"—#$%&amp;()*+,-./:;<=>?@[\\]^_`{|}~\t\n\r«»',
                      lower=True, split=' ', char_level=False)
tokenizer.fit_on_texts([texts])

dist = list(tokenizer.word_counts.items())
print(dist[:10])

data = tokenizer.texts_to_sequences([texts])
res = to_categorical(data[0], num_classes=maxWordsCount)
print(res.shape)
```

# Создаем тензор для обучения

```python
inp_words = 3
n = res.shape[0] - inp_words
X = np.array([res[i:i + inp_words, :] for i in range(n)])
Y = res[inp_words:]
```

# Конфигурация сети (many to one)

```python
model = Sequential()
model.add(Input((inp_words,
                 maxWordsCount)))  # при тренировке в рекуррентные модели keras подается сразу вся последовательность, поэтому в input теперь два числа. 1-длина последовательности, 2-размер OHE
model.add(SimpleRNN(128, activation='tanh'))  # рекуррентный слой на 128 нейронов
model.add(Dense(maxWordsCount, activation='softmax'))
model.summary()
```

# Компиляция модели

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'], optimizer='adam')
```

# Обучение модели

```python
history = model.fit(X, Y, batch_size=32, epochs=50)
```

# Функция предсказания

```python
def buildPhrase(inp_texts, str_len=20):
    res = inp_texts
    print(inp_texts)
    data = tokenizer.texts_to_sequences([inp_texts])[0]
    print(data)
    for i in range(str_len):
        x = to_categorical(data[i: i + inp_words], num_classes=maxWordsCount)  # преобразуем в One-Hot-encoding
        inp = x.reshape(1, inp_words, maxWordsCount)

        pred = model.predict(inp)
        indx = pred.argmax(axis=1)[0]
        data.append(indx)

        res += " " + tokenizer.index_word[indx]  # дописываем строку

    return res
```

# Предсказание

```python
res = buildPhrase("в поисках шести рублей")
print(res)
```
