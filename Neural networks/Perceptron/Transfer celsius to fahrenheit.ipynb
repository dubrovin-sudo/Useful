# Backpropogation method

```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
from tensorflow.keras.layers import Dense
```

Train set

```python
c = np.array([-40, -10, 0, 8, 15, 22, 38])
f = np.array([-40, 14, 32, 46, 59, 72, 100])
print(1.8*c + 32)
```

Создаем полносвязную модель

```python
model = keras.Sequential()
```

Конфигурируем сеть (количество нейронов, количество входов, функция активации)

```python
model.add(Dense(units=1, input_shape=(1,), activation='linear'))
```

Компилируем модель (критерий качества, оптимизатор)

```python
model.compile(loss='mean_squared_error', optimizer=keras.optimizers.Adam(0.1))
```

Обучение (входные значения, выходные значения, количество эпох, вывод служебных данных

```python
history = model.fit(c, f, epochs=500, verbose=0)
print("Обучение завершено")
```

```python
print(model.predict([100]))
print(model.get_weights())
```

```python
plt.plot(history.history['loss'])
plt.grid(True)
plt.show()
```
